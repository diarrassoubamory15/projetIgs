```python
poids = int(input("Entrer le poids :"))
taille = float(input("Entrer la taille :"))

def imc(poids , taille):
    imc = poids/taille**2
    if imc<= 18.5:
        print("poids insuffisant")
    elif 18.5<imc<25:
        print("Corpulence normale")
    elif 25<imc<30:
        print("Surpoids")
    elif 30<imc<35:
        print("Obésité moderée")
    elif 35<imc<40:
        print("Obésité sevère")
    elif imc>=40:
        print("Obésité morbide")
    return imc
    
imc(poids, taille)
```

```python
n= int(input("Entrer un nombre : "))
def factoriel(n):
    i = 1
    for valeur in range(0,n):
        i=i*(valeur+1)
    return i
factoriel(n)
```

```python
from requests import get
from bs4 import BeautifulSoup
url = "https://www.gov.uk/search/news-and-communications"
page = get(url)
soup = BeautifulSoup(page.content, 'html.parser')
aclass = soup.find_all("a",class_="govuk-link")
for title in aclass:
    print(title.string)
```
